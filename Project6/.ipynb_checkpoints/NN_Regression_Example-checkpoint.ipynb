# Regression with Deep Neural Networks

## Setup

First, let's make sure this notebook works well in both python 2 and 3, import a few common modules, ensure MatplotLib plots figures inline and prepare a function to save the figures:

```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "ann"

def save_fig(fig_id, tight_layout=True):
    path = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID, fig_id + ".png")
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format='png', dpi=300)
```

```python
import tensorflow as tf
```

## Create Data

```python
# generate x's, y's for function w/ domain [1,10]
x_s = np.linspace(0,10,1000)
y_s = -1*(np.sin(2*x_s) + x_s**1.2 - x_s**1.25 + np.cos(3*x_s))
# normalize y's to range [0,1]
y_s = (y_s - np.min(y_s))/(np.max(y_s) - np.min(y_s)) #min val = to 0, max = 1
```

```python
plt.plot(x_s,y_s,'b')
```

```python
print(x_s.shape)
# make x_s into a column vector for TF
x_s = np.reshape(x_s,[-1,1])
# shuffle - then split into train and test
rnd_inds = np.arange(len(x_s)) #array xs long
np.random.shuffle(rnd_inds) # xs shuffled
print(x_s.shape)
print(y_s.shape)
# print(rnd_inds)
x_s = x_s[rnd_inds,:] # reorder xs
y_s = y_s[rnd_inds] # ^ + corresponding ys

X_train = x_s[:900] # everything before 900
y_train = y_s[:900]
X_test = x_s[900:] # everything after 900 –––> 100 datapoints
y_test = y_s[900:]

plt.plot(X_train,y_train,'r.') #plot train set
plt.plot(X_test,y_test,'bs') # plot test set
```

```python
# take validation set from training data
# valid used to check quality of test set 
X_valid, X_train = X_train[:100], X_train[100:]
y_valid, y_train = y_train[:100], y_train[100:]
```

```python
plt.plot(X_train, y_train,'r.')
plt.plot(X_valid, y_valid,'gd')
```

## Neural Network

```python
num_epochs = int(1e6) # num training loops
batch_size = 1 # NOTE: ONLY SEEMS TO WORK W/ BATCH SIZE 1

n_inputs = 1
n_outputs = 1
num_nodes_l1 = 10 # 100 doesn't matter
num_nodes_l2 = 10
```

```python
xs = tf.placeholder(tf.float32, shape=(None, n_inputs), name="X") #input values ––> specifying params of fed dict. 
ys = tf.placeholder(tf.float32, shape=(None), name="y") #labels
```

```python
# design network
hidden1 = tf.layers.dense(xs, num_nodes_l1, name="hidden1",
                              activation=tf.nn.relu) #first layer connected to input layer (placeholder)
hidden2 = tf.layers.dense(hidden1, num_nodes_l2, name="hidden2",
                              activation=tf.nn.relu)
outputs = tf.layers.dense(hidden2, n_outputs, name="outputs")
```

```python
cost = tf.reduce_mean(tf.square(outputs-ys)) # mse ... tf.losses.mean_squared_error(labels, predictions)
train = tf.train.AdamOptimizer(0.001).minimize(cost) # optimize (reduce) mse
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(num_epochs):
        # print out progress
        if not(i%(10000)):
            print(i)
        # select random subsets of data to train on (aka batches)
        rnd_idx = np.random.permutation(len(X_train))[:batch_size]
        X_batch = X_train[rnd_idx]
        y_batch = y_train[rnd_idx]
        sess.run(train, feed_dict={xs: X_batch, ys: y_batch}) #feed batch to optimizer
        
    # generate predicted output values for y
    pred_ys = sess.run(outputs, feed_dict={xs:X_test})
```

```python
plt.plot(X_test,y_test,'bd')
plt.plot(X_test,pred_ys,'rd')
```

## Neural Network

```python
reset_graph()
num_epochs = int(1e5) # num training loops
batch_size = 1 # NOTE: ONLY SEEMS TO WORK W/ BATCH SIZE 1

n_inputs = 1
n_outputs = 1
num_nodes_l1 = 100
num_nodes_l2 = 50
```

```python
xs = tf.placeholder(tf.float32, shape=(None, n_inputs), name="X") #input values ––> specifying params of fed dict. 
ys = tf.placeholder(tf.float32, shape=(None), name="y") #labels
```

```python
# design network
hidden1 = tf.layers.dense(xs, num_nodes_l1, name="hidden1",
                              activation=tf.nn.relu) 
#first layer connected to input layer (placeholder)
hidden2 = tf.layers.dense(hidden1, num_nodes_l2, name="hidden2",
                              activation=tf.nn.relu)
outputs = tf.layers.dense(hidden2, n_outputs, name="outputs")
```

```python
cost = tf.reduce_mean(tf.square(outputs-ys)) 
# mse ... tf.losses.mean_squared_error(labels, predictions)
train = tf.train.AdamOptimizer(0.001).minimize(cost) 
# optimize (reduce) mse
```

```python
#number of recorded cost vals
num_record = (int(num_epochs/1000),1)
cost_vals = np.zeros(num_record)
# cost_vals
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    j = 0
    for i in range(num_epochs):
        # print out progress
        if not(i%(10000)):
            print(i)
        # select random subsets of data to train on (aka batches)
        rnd_idx = np.random.permutation(len(X_train))[:batch_size]
        X_batch = X_train[rnd_idx]
        y_batch = y_train[rnd_idx]
        sess.run(train, feed_dict={xs: X_batch, ys: y_batch}) #feed batch to optimizer
        if not(i%1000):
            cost_vals[j] = cost.eval(feed_dict={xs: X_batch, ys: y_batch})
            j += 1
    # generate predicted output values for y
    pred_ys = sess.run(outputs, feed_dict={xs:X_test})
```

```python
plt.plot(X_test,y_test,'bd')
plt.plot(X_test,pred_ys,'rd')
```

```python
#meshgrid... To create a rectangular grid, we need every combination of the x and y points.
#EG)
# x =   0 1 2 3 4        y =   0 0 0 0 0
#       0 1 2 3 4              1 1 1 1 1
#       0 1 2 3 4              2 2 2 2 2
#       0 1 2 3 4              3 3 3 3 3
#       0 1 2 3 4              4 4 4 4 4

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
X_s, y_s = np.meshgrid(x, y) #grid of 100 x 100 points

z_s = X_s**2 + y_s**2 #labels
z_s = (z_s - np.min(z_s))/(np.max(z_s) - np.min(z_s)) #min val = to 0, max = 1

X_s.shape
# z_s.shape
```

```python

# shuffle - then split into train and test
rnd_inds = np.arange(len(points)) #array xs long
np.random.shuffle(rnd_inds) # xs shuffled

points = points[rnd_inds,:] # reorder xs
z_s = z_s[rnd_inds] # ^ + corresponding ys

P_train = points[:9000] # everything before 900
z_train = z_s[:9000]
P_test = points[9000:] # everything after 900 –––> 100 datapoints
z_test = z_s[9000:]

plt.plot(P_train,z_train,'r.') #plot train set
plt.plot(P_test,z_test,'bs') # plot test set
```

```python
reset_graph()

num_epochs = int(1e5) # num training loops
batch_size = 1 # NOTE: ONLY SEEMS TO WORK W/ BATCH SIZE 1

n_inputs = 2
n_outputs = 1
num_nodes_l1 = 10
num_nodes_l2 = 10
```

```python
Ps = tf.placeholder(tf.float32, shape=(None, n_inputs), name="P") #input values ––> specifying params of fed dict. 
zs = tf.placeholder(tf.float32, shape=(None), name="z") #labels

```

```python
# design network
hidden1 = tf.layers.dense(Ps, num_nodes_l1, name="hidden1",
                              activation=tf.nn.relu) #first layer connected to input layer (placeholder)
hidden2 = tf.layers.dense(hidden1, num_nodes_l2, name="hidden2",
                              activation=tf.nn.relu)
outputs = tf.layers.dense(hidden2, n_outputs, name="outputs")
```

```python
cost = tf.reduce_mean(tf.square(outputs-zs)) # mse ... tf.losses.mean_squared_error(labels, predictions)
train = tf.train.AdamOptimizer(0.001).minimize(cost) # optimize (reduce) mse
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(num_epochs):
        # print out progress
        if not(i%(10000)):
            print(i)
        # select random subsets of data to train on (aka batches)
        rnd_idx = np.random.permutation(len(P_train))[:batch_size]
        P_batch = P_train[rnd_idx]
        z_batch = z_train[rnd_idx]
        sess.run(train, feed_dict={Ps: P_batch, zs: z_batch}) #feed batch to optimizer
        
    # generate predicted output values for y
    pred_zs = sess.run(outputs, feed_dict={Ps: P_test})
```
